```python
df1 = pd.read_csv(io.BytesIO(uploaded1["sheet1.csv"]),on_bad_lines='skip', sep=';',skipinitialspace = True)
df2 = pd.read_csv(io.BytesIO(uploaded2["sheet2.csv"]),on_bad_lines='skip', sep=';',skipinitialspace = True)
```

```python
df1.rename(columns = {'Name':'Date'}, inplace = True)
df1.head(10)
```

```python
df2.rename(columns = {'Name':'Date'}, inplace = True)
df2.head(10)
```

```python
df1 = df1.set_index('Date') # run only once
df2 = df2.set_index('Date')
```

```python
array1 = df1.to_numpy()
```

```python
array2 = df2.to_numpy()
```
