# Calculating Weights

```python
#Formula to calculate weights
weights_bookvalue = df_bookvalue.loc[:,df_bookvalue.select_dtypes(exclude='object').columns]=df_bookvalue.select_dtypes(exclude='object').div(df_bookvalue.select_dtypes(exclude='object').sum(1),0)
weights_bookvalue

weights_grossdiv = df_grossdiv.loc[:,df_grossdiv.select_dtypes(exclude='object').columns]=df_grossdiv.select_dtypes(exclude='object').div(df_grossdiv.select_dtypes(exclude='object').sum(1),0)
weights_grossdiv

weights_totemp = df_totemp.loc[:,df_totemp.select_dtypes(exclude='object').columns]=df_totemp.select_dtypes(exclude='object').div(df_totemp.select_dtypes(exclude='object').sum(1),0)
weights_totemp

weights_totrev = df_totrev.loc[:,df_totrev.select_dtypes(exclude='object').columns]=df_totrev.select_dtypes(exclude='object').div(df_totrev.select_dtypes(exclude='object').sum(1),0)
weights_totrev
```

```python
#weights work!
sommauno = weights_totemp.sum(axis = 1)
sommauno.head(5)
```

# Data Cleaning of Weights

```python
# loading Book Value Weights
url5 = 'https://raw.githubusercontent.com/giorgioterranova/QARM/main/weights_bookvalue.csv'
e_weights_bookvalue = pd.read_csv(url5, on_bad_lines='skip', sep=';',decimal=',')

# loading Gross Dividends Weights
url6 = 'https://raw.githubusercontent.com/giorgioterranova/QARM/main/weights_grossdiv.csv'
e_weights_grossdiv = pd.read_csv(url6, on_bad_lines='skip', sep=';',decimal=',')

# loading Total Employment Weights
url7 = 'https://raw.githubusercontent.com/giorgioterranova/QARM/main/weights_totemp.csv'
e_weights_totemp = pd.read_csv(url7, on_bad_lines='skip', sep=';')

#loading Total Revenue Weights

url8 = 'https://raw.githubusercontent.com/giorgioterranova/QARM/main/weights_totrev.csv'
e_weights_totrev = pd.read_csv(url8, on_bad_lines='skip', sep=';')
```

```python
#dropping long name of companies
e_weights_bookvalue1 = e_weights_bookvalue.drop(e_weights_bookvalue.columns[1], axis=1)
e_weights_grossdiv1 = e_weights_grossdiv.drop(e_weights_grossdiv.columns[1], axis=1)
e_weights_totemp1 = e_weights_totemp .drop(e_weights_totemp .columns[1], axis=1)
e_weights_totrev1 = e_weights_totrev.drop(e_weights_totrev.columns[1], axis=1)
```

```python
#setting date as index - only run once!
e_weights_bookvalue1 = e_weights_bookvalue1.set_index('Date')
e_weights_bookvalue1 = e_weights_bookvalue1.transpose()
e_weights_bookvalue2 = e_weights_bookvalue1.transpose()

e_weights_grossdiv1 = e_weights_grossdiv1.set_index('Date')
e_weights_grossdiv1 = e_weights_grossdiv1.transpose()
e_weights_grossdiv2 = e_weights_grossdiv1.transpose()

e_weights_totemp1 = e_weights_totemp1.set_index('Date')
e_weights_totemp1 = e_weights_totemp1.transpose()
e_weights_totemp2 = e_weights_totemp1.transpose()

e_weights_totrev1 = e_weights_totrev1.set_index('Date')
e_weights_totrev1 = e_weights_totrev1.transpose()
e_weights_totrev2 = e_weights_totrev1.transpose()
e_weights_totrev2
```
